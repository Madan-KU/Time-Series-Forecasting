```python
data = [4000, 3950, 3900, 3888]
# data_r = data
num_scenarios = 1000
prev_records_parameter= 3
forecast_parameter= 0.2

def gen_random_number(prev_records):
    return np.random.uniform(low=0.1*prev_records[1], high=prev_records[2])

scenarios = []
max_periods = 60 #maximum periods to simulation

for i in range(num_scenarios):
    forecast = []
    j = 0
    while j < max_periods:

        random_number = gen_random_number(data[-prev_records_parameter:]) #Added parameters for testing *****

        forecast_value = max(data[-1] - forecast_parameter*random_number, 0) # subtract random number from previous value
        
        forecast.append(forecast_value)
        data.append(forecast_value)

        # Checking forecast for convergence
        if len(forecast) > 10 and np.std(forecast[-10:]) < 0.05*np.mean(forecast[-10:]):
            break
        
        j += 1
        
    scenarios.append(forecast)
    data = [4000, 3950, 3900, 3888] # Reset data 

mean_forecast = np.mean(scenarios, axis=0) #For Plotting Mean Forecast line

#Plotting Scenarios
fig, ax = plt.subplots(figsize=(12, 8))#, dpi=300)
for i in range(num_scenarios):
    ax.plot(range(1, len(scenarios[i])+1), scenarios[i], alpha=0.05)
ax.plot(range(1, len(mean_forecast)+1), mean_forecast, linewidth=2, color="b",label=f'Mean Forecast of all {num_scenarios} Scenarios')

ax.axhline(y=0, linestyle="--", color="r")
ax.set_xlabel("Period")
ax.set_ylabel("Value")
ax.set_title(f" Montecarlo Simulation")
ax.legend()

plt.show()
```
